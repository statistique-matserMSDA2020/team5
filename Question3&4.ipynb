# Question 4: Methode de Monte Carlo

```python
#PROJET A RENDRE QUESTION 4
##Abdoulahi Mbengue
#Seyni Kaire
#Khadim Cisse
import random         
from math import sqrt
from matplotlib.pyplot import*
import matplotlib.pyplot as plt
from scipy.stats import uniform

#permet d'importer toutes les variables aleatoires disponible

def integrale_mc(f, a, b, n):   
    somme = 0                   
    for i in range(0, n):       
        x = random.uniform(a, b)
        somme += f(x)
        
    return somme / n

# On estime la valeur de l'integrale de I2.
a=integrale_mc(lambda x: sqrt(1-x**2),0,1,10000)
print("une estimation de I2 nous donne",a)

##EVOLUTTION GRAPHIQUE DE CETTE ESTIMATION LORSQUE n VARIE

from math import*
from random import*
from matplotlib.pyplot import*

n=int(input("donner la valeur de n"))
r=0

for i in range(n):
    x=random() 
    y=random()
    if x**2+y**2<1: #Vérification de sa position par rapport au cercle de rayon 1
        plt.plot(x,y,'ro') # à l'intérieur
        r=r+1 # on le comptabilise
    else:
        plt.plot(x,y,'bo') #à l'extérieur


p=(r/n) # Calcul de l'approximation de l'aire obtenue

plt.axis([0,1,0,1]) # affichage des résultats de l'expérience
plt.text(0.2,0.2, 'Aire='+str(float(p)),size=30)
plt.show()
```
